```python
import seaborn as sns
def plot(df,column):
  #distplot
  plt.figure(figsize=(15,4))
  plt.subplot(1,3,1)
  sns.distplot(df[column])
  plt.title("distplot for"+" "+column)

  #histogram plot

  plt.subplot(1,3,2)
  sns.histplot(df, x= column, kde= True, bins=30,color="salmon")
  plt.title("histogram plot for"+" "+column)
```

SKEWNESS BEFORE IQR TEST

```python
skew=[]
for col in cont_cols:
    skew.append(df[col].skew())
```

```python

# Apply IQR method to treat outliers
for col in cont_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("âœ… Outliers handled using IQR method.")
```

```python
def evaluate_model(y_true, y_pred, model_name):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
    import numpy as np

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"ðŸ“Š {model_name} Performance:")
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"RÂ² Score: {r2:.2f}")
    print("-" * 40)

# Evaluate both models
evaluate_model(y_test, et_preds, "ExtraTrees Regressor")
evaluate_model(y_test, rf_predict, "RandomForestRegressor")
evaluate_model(y_test, lr_predict, "LinearRegression")
```
